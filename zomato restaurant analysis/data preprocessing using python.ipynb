# Insights


```python
# Which country has the least of the zomato restaurants, top 3?
data.groupby("Country")['RestaurantID'].count().sort_values(ascending=True).head(3)
```

```python
# Which zomato restaurant is the most expensive restaurant overall?
# Decided based on the "average cost for two" column
data.groupby("RestaurantName")['Average_Cost_for_two'].mean().sort_values(ascending=False).head(5)
```

```python
# Which zomato restaurant is the cheapest restaurant overall?
# Decided based on the "average cost for two" column
filtered_data=data[data['Average_Cost_for_two']!=0]
filtered_data.groupby("RestaurantName")['Average_Cost_for_two'].mean().sort_values(ascending=True).head(3)
```

```python
# Which country has the average price range as the most cheapest?
data.groupby("Country")['Price_range'].mean().sort_values(ascending=True).head(5)
```

```python
# Which country has the average price range as the most expensive?
data.groupby("Country")['Price_range'].mean().sort_values(ascending=False).head(5)
```

```python
# Which are the five most expensive restaurants in most expensive country?
qatar_restaurants=data[data['Country']=='Qatar']
top_5_most_expensive_restaurants=qatar_restaurants.groupby('RestaurantName')['Average_Cost_for_two'].mean().sort_values(ascending=False).head(5)
top_5_most_expensive_restaurants
```

```python
# Which are the five most cheapest restaurants in most expensive country?
qatar_restaurants=data[data['Country']=='Qatar']
top_5_most_cheapest_restaurants=qatar_restaurants.groupby('RestaurantName')['Average_Cost_for_two'].mean().sort_values(ascending=True).head(5)
top_5_most_cheapest_restaurants
```

```python
# Which are the five most cheapest restaurants in the most cheapest country?
indian_restaurants=data[(data['Country']=='India') & (data['Average_Cost_for_two']!=0)]
top_5_most_cheapest_restaurants_india=indian_restaurants.groupby('RestaurantName')['Average_Cost_for_two'].mean().sort_values(ascending=True).head(5)
top_5_most_cheapest_restaurants_india
```

```python
# Which are the five most expensive restaurants in the most cheapest country?
indian_restaurants=data[(data['Country']=='India')]
top_5_most_expensive_restaurants_india=indian_restaurants.groupby('RestaurantName')['Average_Cost_for_two'].mean().sort_values(ascending=False).head(5)
top_5_most_expensive_restaurants_india
```

```python
# Which country restaurants have the least average ratings?
data.groupby("Country")['Rating'].mean().sort_values(ascending=True).head(1)
```

```python
# Which country restaurants have the highest average ratings?
data.groupby("Country")['Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Which city has the highest number of restaurants left delivering?
left_online_delivery=data[data['Has_Online_delivery']=='No']
left_online_delivery.groupby("City")['RestaurantID'].count().sort_values(ascending=False).head()
```

```python
# Which city zomato restaurants has the most table bookings?
table_booking_restaurants=data[data['Has_Table_booking']=='Yes']
table_booking_restaurants.groupby("City")['RestaurantID'].count().sort_values(ascending=False).head()
```
